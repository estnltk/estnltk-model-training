## VabamorfWithBertTagger problem on balanced_and_reference_corpus

This problem appeared when tagging block 10,0 and text with index 310 caused an error.

The error is caused in BertMorphTagger by assert after new morph layer is created and morph layer doesn't have the same number of words as 'words' layer.


bert_morph_tagger2.py: 331-332

        assert len(morph_layer) == len(words_layer), \
        f"Failed to rewrite '{morph_layer.name}' layer tokens to '{words_layer.name}' layer words: {len(morph_layer)} != {len(words_layer)}"
