Mida teeb: Filtreerib välja read, kus on kaks tõeväärtuse tunnust Trued ja jätab neist vaid ühe Trueks

^ tähendus:  <br/>
0^0 = 0 <br/>
1^1 = 0  <br/> 
1^0 = 1 <br/> 
0^1 = 1

Kui  (Blood = True ja SerPla = True) = True, siis Blood = True ^ True = False <br/>
Kui  (Blood = False ja SerPla = False) = False, siis Blood = False ^ False = False <br/>
Kui (Blood = False ja SerPla = True) = False, siis Blood = False ^ False = False <br/>
Kui (Blood = True ja Serpla = False) = False, siis Blood = False ^ True = True <br/>

Read 1-6 Blood jääb tõeseks vaid juhul kui Blood = True ja Serpla = False <br/>
Rida 7 Plasma jääb tõeseks vaid juhul kui Plasma = True ja Urine = False <br/>
Rida 8 SerPla jääb tõeseks vaid juhul kui SerPla = True ja Urine = False <br/>
Rida 9 Serum jääb tõeseks vaid juhul kui Serum = True ja Urine = False <br/>


```python
    df.Blood ^= (df.Blood & df.SerPla)
    df.Blood ^= (df.Blood & df.Serum)
    df.Blood ^= (df.Blood & df.Plasma)
    df.Blood ^= (df.Blood & df.CapillaryBlood)
    df.Blood ^= (df.Blood & df.CSF)
    df.Blood ^= (df.Urine & df.Blood)
    df.Plasma ^= (df.Urine & df.Plasma)
    df.SerPla ^= (df.Urine & df.SerPla)
    df.Serum ^= (df.Urine & df.Serum)
```
