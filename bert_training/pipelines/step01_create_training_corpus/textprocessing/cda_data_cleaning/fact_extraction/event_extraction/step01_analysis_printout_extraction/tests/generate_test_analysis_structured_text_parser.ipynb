```python
for file in input_files:
    
    text_type = file[0:5]
    file_nr = file[11]
    
    print('*'*100, '\nFile', file,  ', type', text_type, ', file nr', file_nr)
    type_parser = eval(text_type.capitalize() + 'Parser()')
    
    with open('data/' + file) as f:
        text_raw = f.read()
        
    print('Text raw:\n', text_raw)
    output_df = type_parser(text_raw, '123456') #eval(text_type + "_parser")
    print('Output df:\n')
    display(output_df)
    output_filename = 'data/{text_type}_output{nr}.csv'.format(text_type=text_type, nr=file_nr)
    output_df.to_csv(output_filename)
```
