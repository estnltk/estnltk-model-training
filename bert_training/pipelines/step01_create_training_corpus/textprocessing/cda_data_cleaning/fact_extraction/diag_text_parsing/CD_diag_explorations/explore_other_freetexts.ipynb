```python
conn = psycopg2.connect(connection_string1)
cur = conn.cursor()
```

```python
tagger = CancerStageTagger()
```

```python
cur.execute('''select a.epi_id,  a.raw_text, a.table, a.field, d.diag_code from 
work.run_201910192339_extracted_texts as a
join work.runfull201902071713_diagnosis as d
      on a.epi_id = d.epi_id
      where d.diag_code like '%C%' or d.diag_code like '%D0%'
      or d.diag_code like '%D1%' or d.diag_code like '%D2%' or d.diag_code like '%D3%' or d.diag_code like '%D4%'
      ;''')
```

```python
cd_texts = cur.fetchall()
```

```python
tagged_texts = []
for t in tqdm(cd_texts):
    text = Text(t[1])
    tagger.tag(text)
    if len(text.stages) > 0:
        tagged_texts.append(text)
```
