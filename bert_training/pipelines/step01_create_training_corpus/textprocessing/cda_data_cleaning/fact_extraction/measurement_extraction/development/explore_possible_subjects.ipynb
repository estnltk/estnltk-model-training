```python
import csv
import regex as re
from collections import Counter
import sys

csv.field_size_limit(sys.maxsize)
```

```python
exp = re.compile('([^ ]* )([pP]ikkus|PIKKUS|[Kk]aal|KAAL|[Kk]asv|KASV)')
comp = re.compile('([^ ]*)([pP]ikkus|PIKKUS|[Kk]aal|KAAL|[Kk]asv|KASV)')
```

```python
#exp_counter = Counter()
#comp_counter = Counter()
with open("../../../../../egv/adr_2018/all_text_fields_base201805101622texts.csv", "r") as fin:
    reader = csv.reader(fin)
    for idx, row in enumerate(reader):
        if idx%2 != 0:
            row_exps = exp.findall(row[0])
            row_comps = comp.findall(row[0])
            for e in row_exps:
                exp_counter[e[0]] += 1
            for c in row_comps:
                comp_counter[c[0]] += 1

        if idx%1000 == 0:
            print(idx)
```
