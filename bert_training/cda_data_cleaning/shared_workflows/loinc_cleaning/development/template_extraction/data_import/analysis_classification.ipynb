```python
import os
import sys
import pathlib
import pandas as pd
sys.path
sys.path.append('/home/anneott/Repos/cda-data-cleaning/analysis_data_cleaning/LOINC_cleaning')

try:
    __file__
except:
    __file__ = os.path.abspath('.')

print(os.path.abspath('.'))
directory = pathlib.Path(__file__).parent.as_posix()
columns = ['analysis_name', 'parameter_code', 'parameter_name', 'unit', 'count']

#blood.csv ja urine.csv on genereeritud data_import/initial_analysis_seperation.ipynb põhjal
#võetakse ainult kaks analyysi tyypi
blood = pd.DataFrame.from_csv(os.path.join(directory, 'template_extraction/raw_data/blood.csv'))[columns]
urine = pd.DataFrame.from_csv(os.path.join(directory, 'template_extraction/raw_data/urine.csv'))[columns]

blood['system_type'] = 'blood'
urine['system_type'] = 'urine'
  
df = pd.concat([blood, urine])
df = df[['analysis_name', 'system_type']]
df.drop_duplicates().to_csv(os.path.join(directory, 'template_extraction/raw_data/analysis_classified.csv'))
df.drop_duplicates()
```
