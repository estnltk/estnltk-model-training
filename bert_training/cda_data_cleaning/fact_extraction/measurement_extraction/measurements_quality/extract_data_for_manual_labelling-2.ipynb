```python
def find_measurements(obj, collection):
    measurements = []
    for idx, text in enumerate(collection.select(layers = ['measurements'])):
        if idx%10000 == 0:
            print(idx)
        for measurement in text[1].measurements:
            if obj in measurement.REGEX_TYPE:
                text_start = measurement.start - 50
                text_end = measurement.end + 50
                if text_start < 0:
                    text_start = 0
                if text_end > len(text[1].text):
                    text_end = len(text[1].text)
                context = text[1].text[text_start:text_end].replace(measurement.text, '### ' + measurement.text + ' ###') 
                measurements.append((idx, context, measurement))
    return measurements          
```

```python
def get_rows(regex_type, cur):
    cur.execute("""select text_id, span_nr, span_start, span_end, parameter_name, value, unit, min, max,  
date, REGEX_TYPE, epi_id, event_text from work.run_201905311256__measurements_with_text where REGEX_TYPE = %s;""", (regex_type,))
    drug_rows = cur.fetchall()
    shuffle(drug_rows)
    rows_to_label = []

    for row in drug_rows[:100]:
        new_row = []
        new_row.append(row[0])
        new_row.append(row[1])
        new_row.append(row[-3])
        new_row.append(row[5])
        new_row.append(row[6])

        start1 = row[2]
        end1 = row[3]
        start = row[2]-50
        end = row[3]+50
        if start < 0:
            start = 0
        if end > len(row[-1]) :
            end = len(row[-1])
        context = row[-1][start:start1] + ' #### ' + row[-1][start1:end1] + ' ####' + row[-1][end1:end+1]
        new_row.append(context)
        rows_to_label.append(new_row)
    return rows_to_label    
```

```python
headers = ['text_id', 'span_nr', 'OBJ', 'VALUE', 'UNIT', 'context']
rows_to_write = []
rows_to_write.append(headers)
for name in obj_names:
    print(name)
    rows_to_label = get_rows(name, cur)
    rows_to_write += rows_to_label
```

```python
with open("to_label/first_set.csv", "w") as fout:
    writer = csv.writer(fout)
    for row in rows_to_write:
        writer.writerow(row)
```
