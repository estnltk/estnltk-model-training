```python
mapping = {(('Ühendriigid',),'LOC'): 1,
 (('Leedusse',),'LOC'): 2,
 (('Leedu', 'välisministeeriumi'),'LOC'): 3,
 (('Interfaxile',),'LOC'): 4,
 (('Bill', 'Clinton'),'LOC'): 5, (('Leedu','siseministeeriumi'), 'LOC'): 6}
```

```python
a = pd.Series(mapping).reset_index()
```

```python
a.columns = ['entity','label','count']
```

```python
a
#pd.read_csv(ner_counts.csv)
```

```python
singletons_map = {}
last_word_map = {}
first_words_map = {}

for row in a.itertuples():
    if len(row.entity) == 1:
        curr_map = singletons_map.get(row.entity,{})
        curr_map[row.label] = row.count
        singletons_map[row.entity] = curr_map
    else:
        last_word = row.entity[-1]
        curr_map = last_word_map.get(last_word,{})
        curr_count = curr_map.get(row.label,0) 
        curr_map[row.label] = curr_count + row.count
        last_word_map[last_word] = curr_map
        
        first_words = row.entity[:-1]
        curr_map = first_words_map.get(first_words,{})
        curr_count = curr_map.get(row.label,0) 
        curr_map[row.label] = row.count + curr_count
        first_words_map[first_words] = curr_map
        
```
