```python
from estnltk import Text
from estnltk.storage.postgres import PostgresStorage

pgpass_file='pgp_etc.txt'

schema='estonian_text_corpora'

role='estonian_text_corpora_read'

storage = PostgresStorage(pgpass_file=pgpass_file, schema=schema, role=role)

storage
```

```python
test_collection = storage.get_collection('koondkorpus_base_subset_of_5000_v2')
```

```python
morph_layer = 'v166_morph_analysis'
words_layer = 'v166_words'
sentences_layer = 'v166_sentences'
test_collection.selected_layers = [morph_layer,words_layer,sentences_layer]
```
